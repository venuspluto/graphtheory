```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G=nx.random_geometric_graph(200,0.2)
```

```python
plt.figure(figsize=(20,15))
nx.draw_networkx(G,with_labels=False, node_size=50)
plt.show()
```

```python
C=nx.closeness_centrality(G)
```

```python
C
```

```python
C[0]
```

```python
C[2]
```

```python
C1={}
C2={}
C3={}
for i in range(0,len(C)):
    if (C[i]<=0.3):
        C1[i]=C[i]
    elif (C[i]<=0.35):
        C2[i]=C[i]
    else:
        C3[i]=C[i]
```

```python
C3
```

```python
L=nx.spring_layout(G)
plt.figure(figsize=(20,15))
nx.draw_networkx_nodes(G, pos=L,nodelist=C1,node_color='g',node_size=100)
nx.draw_networkx_nodes(G, pos=L,nodelist=C2,node_color='b',node_size=100)
nx.draw_networkx_nodes(G, pos=L,nodelist=C3,node_color='r',node_size=100)
nx.draw_networkx_edges(G, pos=L,width=1)
plt.show()
```

```python
B=nx.betweenness_centrality(G)
```

```python
B
```

```python
B1={}
B2={}
B3={}
for i in range(0,len(B)):
    if (B[i]<=0.02):
        B1[i]=B[i]
    elif (B[i]<=0.03):
        B2[i]=B[i]
    else:
        B3[i]=B[i]
```

```python
B3
```

```python
L=nx.spring_layout(G)
plt.figure(figsize=(20,15))
nx.draw_networkx_nodes(G, pos=L,nodelist=B1,node_color='g',node_size=100)
nx.draw_networkx_nodes(G, pos=L,nodelist=B2,node_color='b',node_size=100)
nx.draw_networkx_nodes(G, pos=L,nodelist=B3,node_color='r',node_size=100)
nx.draw_networkx_edges(G, pos=L,width=1)
plt.show()
```
