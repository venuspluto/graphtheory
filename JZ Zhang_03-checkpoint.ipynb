```python
import networkx as nx
import matplotlib.pyplot as plt
import random as random 
```

```python
G=nx.erdos_renyi_graph(1000,0.1)
```

```python
degree=G.degree()
```

```python
degree
```

```python
degree_values = sorted(set(degree.values()))
```

```python
degree_values
```

```python
histogram = [degree.values().count(i)/float(nx.number_of_nodes(G))*100 for i in degree_values]                  
```

```python
histogram
```

```python
plt.bar(degree_values,histogram)
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
plt.plot(degree_values,histogram,'o')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.show()
```

```python
plt.plot(degree_values,histogram,'o')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
