## import networkx as nx
import matplotlib.pyplot as plt

# Adjacency matrix and Incidence matrix

```python
G=nx.Graph()
G.add_edges_from([('a','b'),('a','c'),('b','c'),('b','d'),('c','d'),('e','f'),('e','g'),('f','g'),('d','e'),('b','f')\
                 ('c','g')])
```

```python
#nx.draw_networkx(G.width=3.6, node_color='r', node_size=1000)
nx.draw_networkx(G)
plt.show()
```

```python
A = nx.adjacency_matrix(G)
print(A.todense())
```

```python
#vertex ordering used for the adjacency matrix
G.nodes()
```

```python
#edges ordering used for the incidence matrix
G.edges()
```

```python
A2 = A*A
print(A2.todense())
```

```python
M=nx.incidence_matrix(G)
print(M. todense())
```
