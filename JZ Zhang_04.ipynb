```python
M=data.corr()
```

```python
M
```

```python
C=M.as_matrix()
```

```python
C
```

```python
my_list =[]
```

```python
n=len(C)
for i in range(0,n):
    for j in range(0,n):
        if j>i:
            if abs(C[i][j])>0.5:
                my_list.append((i,j))
```

```python
my_list
```

```python
G=nx.from_edgelist(my_list)
```

```python
nx.draw_networkx(G)
plt.show()
```
